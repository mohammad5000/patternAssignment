# **libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

Read and show the image

```python
img = cv2.imread("/content/gray.jpg",0)
res = cv2.resize(img, (640,480), interpolation=cv2.INTER_LINEAR)
cv2_imshow(res)
```




**Feature Extraction**

**Spatial  domain Feature**

```python
averageBlur = cv2.blur(res, (5, 5))
cv2_imshow(averageBlur)
```

```python
contrast = cv2.convertScaleAbs(res, alpha=0.3)

cv2_imshow(contrast)
```

**Segmentation**

**blurred img Adaptive threshold **

```python
averageBlur = cv2.cvtColor(averageBlur, cv2.COLOR_BGR2GRAY)
img_thresh_mean_c = cv2.adaptiveThreshold(averageBlur, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
cv2_imshow(img_thresh_mean_c)
```

**Blurred image otsu's threshhold**

```python
t,img_otsu_binary = cv2.threshold(averageBlur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2_imshow(img_otsu_binary)
```

**Original img otsu's threshold**

```python
t,img_otsu_binary = cv2.threshold(res,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
cv2_imshow(img_otsu_binary)
```
